### **Historical data on Storms**
### **Introduction**
Hurricanes, tropical storms are the two most powerful forms of weather phenomenon. Understanding how they act — their wind speed, pressure systems, seasonality, and geographic distribution — is of the highest priority to improve forecasting and preparedness planning.

We analyze the historical storm records in the dplyr::storms data set that cover extensive information of thousands of named storms. Utilizing data visualization techniques in R with packages like ggplot2 and dplyr, we explore trends with respect to storm intensity, frequency, direction, and time trends.

**Our main objectives are:**
*   To analyze the average wind speed per year
*   To identify the most powerful storms and most recurring storm names.
*   To examine seasonal patterns by comparing storms across months and years.






Through the use of a combination of scatter plots, bar charts, time series, and density plots, we aim to understand the patterns of storms and how their characteristics have evolved over time.



### **Average windspeed of storms per year**

```r
storms %>%
  group_by(year) %>%
  summarize(avg_wind = mean(wind, na.rm = TRUE)) %>%
  ggplot(aes(x = year, y = avg_wind)) +
  geom_line(color = "green", linewidth = 1.2) +
  labs(title = "Average Wind Speed of Storms per Year",
       x = "Year",
       y = "Average Wind Speed (knots)") +
  theme_minimal()
```

### **Top strongest storms by windspeed**

```r
storms %>%
  arrange(desc(wind)) %>%
  slice_head(n = 10) %>%
  ggplot(aes(x = reorder(name, wind), y = wind)) +
  geom_col(fill = "red") +
  coord_flip() +
  labs(title = "Top 10 Strongest Storms by Wind Speed",
       x = "Storm Name",
       y = "Wind Speed (knots)") +
  theme_minimal()
```

### **Number of storms by month**

```r
storms %>%
  count(month) %>%
  ggplot(aes(x = factor(month), y = n)) +
  geom_col(fill = "purple") +
  labs(title = "Number of Storms by Month",
       x = "Month",
       y = "Number of Storms") +
  theme_minimal()
```
